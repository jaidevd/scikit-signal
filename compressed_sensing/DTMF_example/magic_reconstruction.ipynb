```python
# Script for solving the compressed sensing problem from Cleve Moler's article
# "Magic Reconstruction: Compressed Sensing"
# (http://www.mathworks.in/company/newsletters/articles/clevescorner-compressed-sensing.html)
###############################################################################
# Author: Jaidev Deshpande
# CAUTION: This notebook is intended to be run with the --pylab=inline option
```

```python
# Imports
from sklearn.linear_model.sparse import Lasso
from scipy.fftpack import dct, idct
from scipy.sparse import coo_matrix
```

```python
# Initializing constants and signals
N = 5000
FS = 4e4
M = 500
f1, f2 = 697, 1336 # Pick any two touchtone frequencies
duration = 1./8
t = linspace(0, duration, duration*FS)
f = sin(2*pi*f1*t) + sin(2*pi*f2*t)
f = reshape(f, (len(f),1))
```

```python
# Displaying the test signal
plot(t,f)
title('Original Signal')
show()
```

```python
# Randomly sampling the test signal
k = randint(0,N,(M,))
k = sort(k)
b = f[k]
plot(t,f,'b', t[k],b,'r.')
title('Original Signal with Random Samples')
```

```python
# Reconstructing the signal from the random samples with sklearn.linear_model.Lasso
D = dct(eye(N))
A = D[k,:]
lasso = Lasso(alpha=0.001)
lasso.fit(A,b.reshape((M,)))
```

```python
imshow(dot(A,transpose(A))) # Matrix A should be isotropic, i.e. cov(A) should be the identity matrix.
```

The matrix A should be isotropic, i.e.:
$E(AA^T)=I$
Or, the covariance of A should be the identity matrix.

```python
# Plotting the reconstructed coefficients and the signal
plot(lasso.coef_)
title('IDCT of the Reconstructed Signal')
recons = dct(lasso.coef_.reshape((N,1)),axis=0)
figure()
plot(t,recons)
title('Reconstucted Signal')
```

```python
recons_sparse = coo_matrix(lasso.coef_)
sparsity = 1 - float(recons_sparse.getnnz())/len(lasso.coef_)
print sparsity
```

Thus, the coefficients required to construct the original signal can be stored in a matrix that is 90% sparse!
Optionally, we can play out the audio in both the reconstructed signal and the original signal to verify if 
they are the same.
