```python
# Importsfrom sklearn.linear_model.sparse import Lassofrom scipy.fftpack import dct, idctfrom scipy.sparse import coo_matrix
```

```python
# Initializing constants and signalsN = 5000FS = 4e4M = 500f1, f2 = 697, 1336 # Pick any two touchtone frequenciesduration = 1./8t = linspace(0, duration, duration*FS)f = sin(2*pi*f1*t) + sin(2*pi*f2*t)f = reshape(f, (len(f),1))
```

```python
# Displaying the test signalplot(t,f)title('Original Signal')show()
```

```python
# Randomly sampling the test signalk = randint(0,N,(M,))k = sort(k)b = f[k]plot(t,f,'b', t[k],b,'r.')title('Original Signal with Random Samples')
```

```python
# Reconstructing the signal from the random samples with sklearn.linear_model.LassoD = dct(eye(N))A = D[k,:]lasso = Lasso(alpha=0.001)lasso.fit(A,b.reshape((M,)))
```

```python
# Plotting the reconstructed coefficients and the signalplot(lasso.coef_)title('IDCT of the Reconstructed Signal')recons = dct(lasso.coef_.reshape((N,1)),axis=0)figure()plot(t,recons)title('Reconstucted Signal')
```

```python
recons_sparse = coo_matrix(lasso.coef_)sparsity = 1 - float(recons_sparse.getnnz())/len(lasso.coef_)print sparsity
```

Thus, the coefficients required to construct the original signal can be stored in a matrix that is 90% sparse!Optionally, we can play out the audio in both the reconstructed signal and the original signal to verify if they are the same.
